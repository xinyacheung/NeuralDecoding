```python
import torch
import numpy as np
import imageio
from Model.VideoUNet import VideoUNet
```

```python
device=torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
all_region = {"PFC": (0, 30),  
            "FEF": (30, 60),
            "LIP": (60, 90),
            "MT": (90, 110),
            "IT": (110, 120),
            "V4": (120, 140)}

def mask_function(input, position):
    start, end = all_region[position] 
    input[:,:,start:end] = 0      
    return input
```

```python
experiment_name = "V4"
```

```python
stat_dict=torch.load("Model/checkpoint.pth")
model=VideoUNet(input_shape=[3,32,32],ncell=140,hiddensize=32)
model.load_state_dict(stat_dict)
model=model.eval()
```

```python
test_input=torch.load("data/test_input")
test_input=test_input.float()
batch_size_test=int(test_input.shape[0]/64)
loss_all=0
img_list=[]
test_size=test_input.shape[0]
test_input_mask=mask_function(test_input, experiment_name)
with torch.no_grad():
    for batch_idx in range(0, test_size, batch_size_test): 
        torch.cuda.empty_cache()
        model=model.to(torch.device(device))
        batch_test_mask = test_input_mask[batch_idx:batch_idx + 1].to(torch.device(device))
        test_result_mask = model(batch_test_mask)
        img=torch.mean(test_result_mask,dim=0)
        img=img.permute(1,2,3,0).to(torch.device("cpu")).detach().numpy()
        img_list.append(img)
output_filename="result/"
for i in range (16):
    for j in range (4):
        imgs=img_list[i*4+j]
        name=(output_filename+"Train_stimuli"+str(i)+"_cue"+str(j))
        with imageio.get_writer(name+'.gif', mode='I') as writer:
            for img in imgs:
                writer.append_data(np.array(img*255, dtype=np.uint8))
if len(img_list)==64:
    print("success")
```
